```python
from bs4 import BeautifulSoup
import requests

headers = {"User-Agent": "Mozilla/5.0"}

def retrive_data(die,pg_no):
    print('*'*50, die, '*'*50)
    if pg_no < 2:
        print("started ---------------")
        r = requests.get(f"https://www.icliniq.com/qa/{die}", headers=headers)
        c = r.content
        soup = BeautifulSoup(c, "html.parser")
        return soup, r.status_code
    # elif r.status_code == 404:
    #     break
    else:
        r = requests.get(f"https://www.icliniq.com/qa/{die}?page={pg_no}", headers=headers)
        c = r.content
        soup = BeautifulSoup(c, "html.parser")
        return soup, r.status_code
```

```python
die_list = ['Fever','Anxiety','Diabetes','Asthma','Hypertension','Headache','Heart Attack','Viral Infection','Diarrhea'] #'Bad Throat'
```

```python
total_links = {}

for i in die_list:
    for pg_no in range(1, 20):
        soup,status_code = retrive_data(i.lower(),pg_no)
        if soup != None:
            print("Website pages finished ; (")
            
        link_class = soup.find_all("div", {"class": "background-shadow p-3 mb-4"})
        base_url = 'https://www.icliniq.com/'
        links = []
        for die in link_class:
            topic_name = die.find("a")['href']
            link = base_url + topic_name
            #print(topic_name)
            links.append(link)
            
    total_links[i] = links
```
